## Primeiras Visualizações

```python
import pandas as pd
df = pd.read_csv('dados/monitoramento_tempo.csv')
df.head()
```

```python
df.info()
```

```python
# definido como datetime
df['data'] = pd.to_datetime(df['data'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot da
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'],color='green')

eixo.set_title('Temperatura no momento',fontsize=25,color="white")
eixo.set_ylabel('Temperatura',fontsize=20, color="white")
eixo.set_xlabel('Data',fontsize=20,color="white")
eixo.legend(['Temperatura'],loc = 'lower right',fontsize=15)
```
