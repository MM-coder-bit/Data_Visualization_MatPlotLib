## Customizando nossas visualizações

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
# origem dos dados,dados, cor do grafico, espessura da linha(lw), tipo de tracejado(ls)
eixo.plot(df['data'], df['temperatura'],color='g',lw = 0.7,ls='dotted')

eixo.set_title('Temperatura no momento',fontsize=25,color="white")
eixo.set_ylabel('Temperatura',fontsize=20, color="white")
eixo.set_xlabel('Data',fontsize=20,color="white")
eixo.legend(['Temperatura'],loc = 'lower right',fontsize=15)
```

```python
# com marcador 'o'
import datetime

fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
# origem dos dados,dados, cor do grafico, espessura da linha(lw), tipo de tracejado(ls), marcadores(marker)
eixo.plot(df['data'], df['temperatura'],color='g',lw = 0.7,ls='dotted', marker = 'o')
# definindo um periodo da visualização
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
# configuração
eixo.set_title('Temperatura no momento',fontsize=25,color="white")
eixo.set_ylabel('Temperatura',fontsize=20, color="white")
eixo.set_xlabel('Data',fontsize=20,color="white")
eixo.legend(['Temperatura'],loc = 'lower right',fontsize=15)
eixo.grid(True) # grade 
```

```python
# sem marcador 'o'
import datetime

fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
# origem dos dados,dados, cor do grafico, espessura da linha(lw), tipo de tracejado(ls)
eixo.plot(df['data'], df['temperatura'],color='g',lw = 0.9,ls='dotted')
# definindo um periodo da visualização
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
# configuração
eixo.set_title('Temperatura no momento',fontsize=25,color="white")
eixo.set_ylabel('Temperatura',fontsize=20, color="white")
eixo.set_xlabel('Data',fontsize=20,color="white")
eixo.legend(['Temperatura'],loc = 'lower right',fontsize=15)
eixo.grid(True) # grade 
```

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.plot(df['data'], df['temperatura'], color = 'g')
# definindo um periodo da visualização
eixo.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eixo.set_ylim(270,320)
eixo.set_title('Temperatura de Maio de 2014', fontsize=25, pad=20,color="white") # pad = espaçamento do texto com o gráfico
eixo.legend(['Temperatura'], loc='lower right', fontsize=15) # 
eixo.set_ylabel('Temperatura',fontsize=20, color="white")
eixo.set_xlabel('Data',fontsize=20,color="white")
eixo.grid(True)
# ------------------------------------------------------------------------------------------------------
# Aplicando filtro no gráfico
azul_esq = df['data'] < datetime.datetime(2014,5,1)
azul_dir = df['data'] > datetime.datetime(2014,6,1)

eixo2.plot(df['data'],df['temperatura'],color = 'g')
eixo2.plot(df[azul_esq] ['data'],df[azul_esq]['temperatura'],color = 'b')
eixo2.plot(df[azul_dir] ['data'],df[azul_dir]['temperatura'],color = 'b')
# -----------------------------------------------------------------------------------------------------
# definindo um periodo da visualização
eixo2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo2.set_title('Temperatura em 2014', fontsize=15, pad=2) # pad = espaçamento do texto com o gráfico
eixo2.legend(['Temperatura'], loc='best', fontsize=8) # loc = best, o algoritmo insere no melhor lugar
eixo2.set_ylabel('Temperatura',fontsize=10, color="white")
eixo2.set_xlabel('Data',fontsize=10,color="white")
eixo2.grid(True)
```

## Linhas de restrição e anotações

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

# Dados de temperatura
eixo.plot(df['data'], df['temperatura'], color = 'g')
# Labels
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20,color='white')
eixo.set_xlabel('Data', fontsize=20,color='white')
eixo.set_ylabel('Temperatura', fontsize=20,color='white')
# Legenda no gráfico
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

# x1/2 = idxmax() (de index max) é uma função do Pandas que nos retorna exatamente o índice máximo da coluna
# y1/2 = Valor máximo em Y
x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) - 5

eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, 
              xytext=(x2,y2), arrowprops=dict(facecolor='lightgreen')) # anotação no ponto
# minimo
x3 = df['data'][df['temperatura'].idxmin()]
y3 = min(df['temperatura'])

x4 = df['data'][df['temperatura'].idxmin() - 7000]
y4 = min(df['temperatura']) + 5

eixo.annotate('Minimo', xy=(x3,y3), fontsize=20, 
              xytext=(x4,y4), arrowprops=dict(facecolor='lightgreen')) # anotação no ponto

# tracejado no gráfico
eixo.axhline(max(df['temperatura']), color = 'r', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'r', linestyle='--')
```
